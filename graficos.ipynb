```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('resultados.csv', delimiter=',')
df.sort_values(by='Tiempo_s')
```

```python
plt.clf()
df.sort_values(by='Informantes').plot(x='Informantes', y='Tiempo_s', logy='true')
plt.show()
```
